#JSON FILE READING WITH PANDAS

JSON files are lightweight text-based, human-readable files and can be edited using a text editor, **these files are useful to store simple data structures and objects in a specific format.**

The purpouse of this notebook is to practice JSON file manipulation with Pandas to reach a better comprehension of data through a simple in a  **Data Frame.**



Imagen1.jpg

```python
import requests
import json
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/bttmly/nba/master/data/teams.json'
req=requests.get(url)
print(req)
```

```python
texto=req.text
texto
```

```python
jsondata=json.loads(texto)
jsondata
```

```python
df=pd.DataFrame.from_dict(jsondata)
df
```

```python
df.info()
```
